```python
df['age'] = 2020 - df['year_built'] 
df.head()
```

```python
import geopy.distance
from geopy.distance import lonlat, distance
from tqdm import tqdm
tqdm.pandas()

central_park = (40.785091, -73.968285)
df['distance'] = df.progress_apply(lambda row: geopy.distance.great_circle(central_park, (row.latitude, row.longitude)).km, axis=1)
```

```python
import numpy as np 
df['distance'] = df['distance'].apply(lambda x: np.nan if x >= 100 else x) 
df.head()
```

```python
df_num = df[['sale_price', 'distance', 'res_units', 'comm_units', 'total_units','age']].astype(float)
```

```python
df_num.dtypes
```

```python
corr = df_num.dropna().corr()
cmap = sns.diverging_palette(220,10,as_cmap=True)
sns.heatmap(corr,vmax=.3,center=0,cmap=cmap,square=True, linewidth=.5, cbar_kws={"shrink":.5})
```
